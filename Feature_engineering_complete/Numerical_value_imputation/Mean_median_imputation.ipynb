### Imputation of missing values- Numerical

1. Mean median imputation 


```python
### Importing the necessary libraries 

import pandas as pd
import numpy as np
```

```python
## Loading the dataset

data = pd.read_csv('C:\Feature_engineering_complete\Datasets\House Price.csv')
pd.set_option('display.max_columns',None)
```

```python
data.head()
```

###### We can impute mean or median  for numerical missing values. 

```python
## Simple function to impute nan values 

def impute_nan(data,variable):
    data[variable+'_mean'] = data[variable].fillna(data[variable].mean())
    data[variable+'_median'] = data[variable].fillna(data[variable].median())
```

```python
## Numerical features

numerical_features = [x for x in data.columns if data[x].dtype!='O']
```

```python
data[numerical_features].isnull().sum()
```

```python
## We have LotFrontage  and GarageYrBlt. We will impute the function for both the features. 

impute_nan(data,'LotFrontage')
impute_nan(data,'GarageYrBlt')
```

```python
data.head()
```

```python
print('New_columns_null_values:\n',data.loc[:,'LotFrontage_mean':'GarageYrBlt_median'].isnull().sum())
```

###### Plot to determine if the imputation was meaningful. 
#####  We will drop the null values of the original numerical values and compare the distribution with the new columns

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,12))
plt.subplot(3,3,1)
data['LotFrontage'].dropna().plot.hist()
plt.xlabel('LotFrontage')
plt.subplot(3,3,2)
data['LotFrontage_mean'].plot.hist()
plt.xlabel('LotFrontage_mean')
plt.subplot(3,3,3)
data['LotFrontage_median'].plot.hist()
plt.xlabel('LotFrontage_median')
plt.show()
```

```python
plt.figure(figsize=(16,12))
plt.subplot(3,3,1)
data['GarageYrBlt'].dropna().plot.hist()
plt.xlabel('GarageYrBlt')
plt.subplot(3,3,2)
data['GarageYrBlt_mean'].plot.hist()
plt.xlabel('GarageYrBlt_mean')
plt.subplot(3,3,3)
data['GarageYrBlt_median'].plot.hist()
plt.xlabel('GarageYrBlt_median')
plt.show()
```
