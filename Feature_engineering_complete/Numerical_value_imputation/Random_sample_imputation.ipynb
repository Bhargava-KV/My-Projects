### Imputation of missing values- Numerical

2. Random Sample imputation 


```python
### Importing the necessary libraries 

import pandas as pd
import numpy as np
```

```python
## Loading the dataset

data = pd.read_csv('C:\Feature_engineering_complete\Datasets\House Price.csv')
pd.set_option('display.max_columns',None)
```

```python
data.head()
```

```python
## Numerical features

numerical_features = [x for x in data.columns if data[x].dtype!='O']
```

###### We can impute random sample from the feature for numerical missing values. 

```python
## Simple function to impute nan values 

def impute_nan(data,variable):
    data[variable+'_random'] = data[variable]
    random_sample = data[variable].dropna().sample(data[variable].isnull().sum(),random_state=2)
    ## pandas require indexing 
    random_sample.index= data[data[variable].isnull()].index
    data.loc[data[variable].isnull(),variable+'_random']= random_sample
```

```python
data[numerical_features].isnull().sum()
```

```python
## We will impute nan for LotFrontage  and GarageYrBlt 
```

```python
impute_nan(data,'LotFrontage')
impute_nan(data,'GarageYrBlt')
```

###### Plot to determine if the imputation was meaningful.

```python
plt.figure(figsize=(16,12))
plt.subplot(4,4,1)
data['LotFrontage'].dropna().plot.hist()
plt.xlabel('LotFrontage')
plt.subplot(4,4,2)
data['LotFrontage_random'].plot.hist()
plt.xlabel('LotFrontage_random')
plt.subplot(4,4,3)
data['GarageYrBlt'].plot.hist()
plt.xlabel('GarageYrBlt')
plt.subplot(4,4,4)
data['GarageYrBlt_random'].plot.hist()
plt.xlabel('GarageYrBlt_random')
plt.show()
```
